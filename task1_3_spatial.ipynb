##Section2 : DATA CLEANING + BASE FEATURE ENGINEERING

```python
# considering only the important columns
df_clean = df[[
    "tpep_pickup_datetime",
    "tpep_dropoff_datetime",
    "PULocationID",
    "DOLocationID",
    "trip_distance",
    "fare_amount",
    "total_amount",
    "passenger_count"
]].copy()
```

```python
# 1.4lakh trip count had passenger_count null
df_clean["passenger_count"] = df_clean["passenger_count"].fillna(1)
```

```python
# considering only valid trips
df_clean = df_clean[df_clean["trip_distance"] > 0]
df_clean = df_clean[df_clean["fare_amount"] > 0]
df_clean = df_clean[df_clean["total_amount"] > 0]
df_clean = df_clean[df_clean["passenger_count"] > 0]
```

```python
df_clean["pickup_time"] = pd.to_datetime(df_clean["tpep_pickup_datetime"])
df_clean["drop_time"] = pd.to_datetime(df_clean["tpep_dropoff_datetime"])
```

```python
# data had some irrelavent years
df_clean = df_clean[df_clean["pickup_time"].dt.year == 2024]
```

```python
# creating trip duration feature and filtering the valid trips accordingly
df_clean["trip_duration"] = (
    df_clean["drop_time"] - df_clean["pickup_time"]
).dt.total_seconds() / 60
```

```python
df_clean = df_clean[(df_clean["trip_duration"] > 1) & (df_clean["trip_duration"] < 120)]
```

```python
# more derived feature
df_clean["hour"] = df_clean["pickup_time"].dt.hour
df_clean["day_of_week"] = df_clean["pickup_time"].dt.dayofweek
df_clean["month"] = df_clean["pickup_time"].dt.month
```

```python
df_clean["speed"] = df_clean["trip_distance"] / (df_clean["trip_duration"] / 60)
df_clean["fare_per_km"] = df_clean["fare_amount"] / df_clean["trip_distance"]
```

```python
for col in ["trip_distance", "trip_duration", "fare_amount", "speed"]:
    low = df_clean[col].quantile(0.01)
    high = df_clean[col].quantile(0.99)
    df_clean = df_clean[(df_clean[col] >= low) & (df_clean[col] <= high)]
```

```python
# ============================================================
# TASK 1: SPATIAL FEATURE ENGINEERING + ML MODELS
# ============================================================

import pandas as pd
import numpy as np
import geopandas as gpd
import glob
import warnings
warnings.filterwarnings("ignore")

# ── 1. Load data ──────────────────────────────────────────
files = glob.glob("clean_*.parquet")
df_all = pd.concat([pd.read_parquet(f) for f in files], ignore_index=True)

zones = pd.read_csv("https://d37ci6vzurychx.cloudfront.net/misc/taxi_zone_lookup.csv")
zones = zones[["LocationID", "Borough", "Zone"]]

gdf = gpd.read_file("taxi_zones/taxi_zones.shp")
gdf = gdf[["LocationID", "geometry"]]
zone_info = gdf.merge(zones, on="LocationID", how="left")
zone_info = zone_info[~zone_info["Borough"].isin(["Unknown", "EWR"])]

# ── 2. Build zone-level feature table ─────────────────────
# trip_count is your existing variable — directly from groupby
zone_counts = (
    df_all.groupby("PULocationID")
    .size()
    .reset_index(name="trip_count")
)

pickup_stats = df_all.groupby("PULocationID").agg(
    avg_fare=("fare_amount", "mean"),
    avg_distance=("trip_distance", "mean"),
    avg_speed=("speed", "mean"),
    avg_duration=("trip_duration", "mean"),
    avg_passengers=("passenger_count", "mean")
).reset_index().rename(columns={"PULocationID": "LocationID"})

# Merge trip_count into pickup_stats
pickup_stats = pickup_stats.merge(
    zone_counts.rename(columns={"PULocationID": "LocationID"}),
    on="LocationID", how="left"
)

dropoff_count = df_all.groupby("DOLocationID").size().reset_index(name="dropoff_count")
dropoff_count.rename(columns={"DOLocationID": "LocationID"}, inplace=True)

zone_feat = zone_info.merge(pickup_stats, on="LocationID", how="left")
zone_feat = zone_feat.merge(dropoff_count, on="LocationID", how="left")
zone_feat = zone_feat.fillna(0)

# net flow
zone_feat["net_flow"] = zone_feat["trip_count"] - zone_feat["dropoff_count"]

# centroid coordinates
zone_feat = zone_feat.to_crs(epsg=4326)
zone_feat["centroid_lon"] = zone_feat.geometry.centroid.x
zone_feat["centroid_lat"] = zone_feat.geometry.centroid.y

# zone area in sq km
zone_feat_proj = zone_feat.to_crs(epsg=32618)
zone_feat["area_km2"] = zone_feat_proj.geometry.area / 1e6

# Borough encoding
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
zone_feat["borough_enc"] = le.fit_transform(zone_feat["Borough"].fillna("Unknown"))

print("Zone feature table shape:", zone_feat.shape)
zone_feat[["LocationID", "Zone", "Borough", "trip_count",
           "net_flow", "centroid_lat", "centroid_lon", "area_km2"]].head()
```

```python
# ── 3. Spatial Lag Features (Queen Contiguity) ────────────
from libpysal.weights import Queen
import libpysal

zone_feat_proj = zone_feat.to_crs(epsg=32618).reset_index(drop=True)
w = Queen.from_dataframe(zone_feat_proj, silence_warnings=True)
w.transform = 'r'

# lag_trip_count = weighted avg of neighboring zones' trip_count
zone_feat_proj["lag_trip_count"] = libpysal.weights.lag_spatial(w, zone_feat_proj["trip_count"])
zone_feat_proj["lag_avg_fare"]   = libpysal.weights.lag_spatial(w, zone_feat_proj["avg_fare"])
zone_feat_proj["lag_net_flow"]   = libpysal.weights.lag_spatial(w, zone_feat_proj["net_flow"])

print("Spatial lag features added.")
print(zone_feat_proj[["Zone", "trip_count", "lag_trip_count", "net_flow", "lag_net_flow"]].head(8))
```

```python
# ── 4. Prepare ML Dataset (FIXED) ─────────────────────────

feature_cols = [
    "avg_fare", "avg_distance", "avg_speed", "avg_duration", "avg_passengers",
    "dropoff_count", "net_flow",
    "centroid_lat", "centroid_lon", "area_km2", "borough_enc",
    "lag_trip_count", "lag_avg_fare", "lag_net_flow"
]

target_col = "trip_count"

df_ml = zone_feat_proj[zone_feat_proj["trip_count"] > 0].copy()
df_ml = df_ml.dropna(subset=feature_cols + [target_col])

# --- LOG TRANSFORM (VERY IMPORTANT) ---
y = np.log1p(df_ml[target_col].values)
X = df_ml[feature_cols].values

# --- SPATIAL SPLIT (by borough) ---
from sklearn.model_selection import GroupShuffleSplit

groups = df_ml["borough_enc"]

gss = GroupShuffleSplit(n_splits=1, test_size=0.2, random_state=42)
train_idx, test_idx = next(gss.split(X, y, groups))

X_train, X_test = X[train_idx], X[test_idx]
y_train, y_test = y[train_idx], y[test_idx]

print(f"Training samples: {X_train.shape[0]}, Test samples: {X_test.shape[0]}")
```

```python
# ── 5. Random Forest (FIXED) ─────────────────────────────
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, r2_score, mean_squared_error

rf = RandomForestRegressor(
    n_estimators=200,
    max_depth=10,
    random_state=42,
    n_jobs=-1
)

rf.fit(X_train, y_train)

# Predict (log scale)
y_pred_log = rf.predict(X_test)

# Convert back to original scale
y_test_orig = np.expm1(y_test)
y_pred_orig = np.expm1(y_pred_log)

# Metrics
mae_rf  = mean_absolute_error(y_test_orig, y_pred_orig)
rmse_rf = np.sqrt(mean_squared_error(y_test_orig, y_pred_orig))
r2_rf   = r2_score(y_test_orig, y_pred_orig)

print("=== Random Forest (Fixed) ===")
print(f"MAE:  {mae_rf:.2f} trips")
print(f"RMSE: {rmse_rf:.2f} trips")
print(f"R²:   {r2_rf:.4f}")

# Feature importance
feat_imp = pd.Series(rf.feature_importances_, index=feature_cols).sort_values(ascending=False)
```

```python
import matplotlib.pyplot as plt

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Feature importance
feat_imp.head(10).plot(kind="barh", ax=axes[0], color="steelblue")
axes[0].set_title("Random Forest – Feature Importances")
axes[0].set_xlabel("Importance")
axes[0].invert_yaxis()

# Scatter plot (log-log scale for clarity)
axes[1].scatter(y_test_orig, y_pred_orig, alpha=0.6, edgecolors='k', linewidths=0.4)

# Diagonal line
min_val = min(y_test_orig.min(), y_pred_orig.min())
max_val = max(y_test_orig.max(), y_pred_orig.max())
axes[1].plot([min_val, max_val], [min_val, max_val], 'r--')

axes[1].set_xscale("log")
axes[1].set_yscale("log")

axes[1].set_xlabel("Actual Trip Count")
axes[1].set_ylabel("Predicted Trip Count")
axes[1].set_title(f"RF (Log-Scale): R² = {r2_rf:.3f}")

plt.tight_layout()
plt.show()
```

```python
# ── 6. Model 2: XGBoost (FIXED) ───────────────────────────
from xgboost import XGBRegressor

xgb = XGBRegressor(
    n_estimators=300,
    learning_rate=0.05,
    max_depth=6,
    subsample=0.8,
    colsample_bytree=0.8,
    random_state=42,
    n_jobs=-1,
    verbosity=0
)

xgb.fit(X_train, y_train)

# Predict (log scale)
y_pred_log_xgb = xgb.predict(X_test)

# Convert back
y_pred_xgb = np.expm1(y_pred_log_xgb)

# Metrics
mae_xgb  = mean_absolute_error(y_test_orig, y_pred_xgb)
rmse_xgb = np.sqrt(mean_squared_error(y_test_orig, y_pred_xgb))
r2_xgb   = r2_score(y_test_orig, y_pred_xgb)

print("=== XGBoost (Fixed) ===")
print(f"MAE:  {mae_xgb:.2f} trips")
print(f"RMSE: {rmse_xgb:.2f} trips")
print(f"R²:   {r2_xgb:.4f}")
```

```python
# ── 7. Model 3: Spatial OLS (CLEANED) ─────────────────────
from spreg import OLS
from libpysal.weights import Queen as QueenW

feature_names_ols = [
    "avg_distance",
    "avg_duration",
    "area_km2",
    "borough_enc",
    "lag_trip_count"   # keep ONE spatial signal only
]

df_ols = df_ml.dropna(subset=feature_names_ols + [target_col]).reset_index(drop=True)

X_ols = df_ols[feature_names_ols].values
y_ols = df_ols[[target_col]].values

# Spatial weights
w_ols = QueenW.from_dataframe(
    df_ols.set_geometry(df_ols.geometry),
    silence_warnings=True
)
w_ols.transform = 'r'

ols_model = OLS(
    y_ols, X_ols,
    w=w_ols,
    name_y=target_col,
    name_x=feature_names_ols,
    spat_diag=True,
    moran=True
)

print(ols_model.summary)
```

```python
# ── B. LISA – Local Spatial Hot Spots ─────────────────────
from esda.moran import Moran_Local

lisa = Moran_Local(y_moran, w, seed=42)

sig   = lisa.p_sim < 0.05
quads = lisa.q  # 1=HH, 2=LH, 3=LL, 4=HL

labels    = {1: "High-High (Hot)", 2: "Low-High", 3: "Low-Low (Cold)", 4: "High-Low"}
color_map = {1: "red", 2: "lightblue", 3: "blue", 4: "pink", 0: "lightgrey"}

zone_feat_proj["lisa_q"]   = np.where(sig, quads, 0)
zone_feat_proj["lisa_lab"] = zone_feat_proj["lisa_q"].map({0: "Not Sig", **labels})

fig, ax = plt.subplots(figsize=(14, 14))
for q, color in color_map.items():
    subset = zone_feat_proj[zone_feat_proj["lisa_q"] == q]
    if len(subset):
        subset.plot(ax=ax, color=color, edgecolor="black", linewidth=0.3,
                    label={0: "Not Sig", **labels}.get(q, ""))
ax.legend(loc="lower right", fontsize=10)
ax.set_title("LISA Cluster Map – Taxi Trip Count (Pickup Demand)", fontsize=16)
ax.axis("off")
plt.savefig("lisa_cluster_map_trip_count.png", dpi=150)
plt.show()

hh = zone_feat_proj[zone_feat_proj["lisa_q"] == 1][["Zone", "Borough", "trip_count"]]
print("\nHigh-High Hot Spot Zones:")
print(hh.sort_values("trip_count", ascending=False).to_string(index=False))
```

```python
# ── E. Borough Collocation Matrix (MEMORY SAFE) ───────────
import pandas as pd
import numpy as np
import glob
import seaborn as sns
import matplotlib.pyplot as plt

files = glob.glob("clean_*.parquet")

borough_pairs = {}

# Load lookup once
zones = pd.read_csv("https://d37ci6vzurychx.cloudfront.net/misc/taxi_zone_lookup.csv")
zone_to_borough = dict(zip(zones["LocationID"], zones["Borough"]))

for f in files:
    df = pd.read_parquet(f, columns=["PULocationID", "DOLocationID"])

    # Map boroughs (lightweight)
    df["PU_Borough"] = df["PULocationID"].map(zone_to_borough)
    df["DO_Borough"] = df["DOLocationID"].map(zone_to_borough)

    # Remove unwanted
    df = df[
        (~df["PU_Borough"].isin(["Unknown", "EWR"])) &
        (~df["DO_Borough"].isin(["Unknown", "EWR"]))
    ]

    # Count borough pairs
    counts = df.groupby(["PU_Borough", "DO_Borough"]).size()

    for (pu, do), v in counts.items():
        borough_pairs[(pu, do)] = borough_pairs.get((pu, do), 0) + v

    del df  # free memory

# Convert to DataFrame
borough_od = pd.DataFrame(
    [(k[0], k[1], v) for k, v in borough_pairs.items()],
    columns=["PU_Borough", "DO_Borough", "trip_count"]
)

# Pivot
borough_pivot = borough_od.pivot(
    index="PU_Borough",
    columns="DO_Borough",
    values="trip_count"
).fillna(0)

# Plot
fig, ax = plt.subplots(figsize=(8, 6))
sns.heatmap(
    np.log1p(borough_pivot),
    annot=True,
    fmt=".1f",
    cmap="Blues",
    linewidths=0.5,
    ax=ax,
    cbar_kws={"label": "log(trip_count + 1)"}
)

ax.set_title("Borough-Level Trip Collocation Matrix", fontsize=13)
ax.set_xlabel("Dropoff Borough")
ax.set_ylabel("Pickup Borough")

plt.tight_layout()
plt.show()
```
